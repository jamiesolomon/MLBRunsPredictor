```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing all seperate dataframes
Teams = pd.read_csv('Teams.csv')
```

### The MLB switched to a 162 game season in 1961, therefore this report will be using this as the cutoff date for relevent data.

## `Teams` dataframe

```python
Teams.info()
```

```python
Teams = Teams.loc[Teams.yearID >= 1901]
```

```python
plt.figure(figsize=(11,13))
pic = sns.heatmap(Teams.set_index('yearID').sort_index().isna().transpose())
```

```python
Teams = Teams.drop('WCWin', axis=1)
```

```python
# Note that there was no World series in 1994 so all of these values are null. 
# This report will drop all data entries in the year 1994
Teams.loc[Teams.isna().any(axis=1)][['yearID', 'WSWin']]
```

```python
# There was a lockout in 1994
Teams = Teams.drop(Teams[Teams['yearID'] == 1994].index)
```

```python
# MLB began tracking caught stealing in 1951.The model has indicated CS as a significant stastic, however there is not enough data
# to determine whether or not 'SF' or 'HBP' are significant. Therefore, this replort will drop the latter 2 columns after computing OPS
Teams = Teams.loc[Teams.yearID >= 1951]
```

```python
# Adding OBS, Slugging %, and OPS
# OBP = (H + BB + HBP)/(AB + BB + HBP + SF)
Teams['OBP'] = (Teams['H'] + Teams['BB'] + Teams['HBP'])/(Teams['AB'] + Teams['BB'] + Teams['HBP'] + Teams['SF'])
Teams['TB'] = Teams['H'] + Teams['2B'] + (Teams['3B'] * 2) + (Teams['HR'] * 3)
Teams['SLG'] = Teams['TB'] / Teams['AB']
Teams['OPS'] = Teams['OBP'] + Teams['SLG']
```

#### Note that OPS data in 1951 is not available. This report will mind the missing data from before 1974 from MLB.com using the library BeautfulSoup

## ------MINING-----

```python
# Adding imports after so it wont run every time i run all lines
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.DataFrame(columns=['Year', 'Team', 'OPS'])

# Loop over each year and each team
for year in range(1951, 1974):
    print(f'\n----------------------------\nMining year {year}\n----------------------------------\n')
    for offset in range(0, 1000, 100):
        # Retrieve the HTML for the current year and offset
        url = f'https://www.mlb.com/stats/team/{year}?offset={offset}&page=1&playerPool=ALL'
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        # Get the number of teams in the current year
        num_teams = len(soup.select('table tr')) - 1

        # Loop over each team in the current year
        for i in range(0, num_teams):
            # Retrieve the data for the current team in the current year
            team_name = soup.select(f'table tr:nth-of-type({i+1}) th a')[0].text.strip()
            team_ops = soup.select(f'table tr:nth-of-type({i+1}) td:nth-of-type(17)')[0].text.strip()
            team_obp = soup.select(f'table tr:nth-of-type({i+1}) td:nth-of-type(15)')[0].text.strip()

            # Add the data to the dataframe
            df = df.append({'Year': year, 'Team': team_name, 'OBP': team_obp, 'OPS': team_ops}, ignore_index=True)
            
# Cleaning up team name using regex
df['Team'] = df['Team'].str.replace(r'([a-z])([A-Z]).*', r'\1') 

# Save the dataframe to a CSV file
df.to_csv('OPS_data.csv', index=False)
```
